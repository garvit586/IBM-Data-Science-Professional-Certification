````python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year

yearly_sales = df.groupby('Year')['Automobile_Sales'].sum()

plt.figure(figsize=(12, 6))
yearly_sales.plot(kind='line', marker='o', color='blue')

plt.title("Yearly Automobile Sales Fluctuations", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Automobile Sales", fontsize=12)
plt.grid(True)
plt.xticks(yearly_sales.index, rotation=45)

plt.tight_layout()
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>
</details>


````python
# Filter data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Group data by Vehicle_Type and calculate average sales for both periods
recession_sales = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
non_recession_sales = non_recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()

# Add a column to indicate the period
recession_sales['Period'] = 'Recession'
non_recession_sales['Period'] = 'Non-Recession'

# Combine the datasets for visualization
combined_sales = pd.concat([recession_sales, non_recession_sales])

# Plot the sales trend comparison using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Vehicle_Type', y='Automobile_Sales', hue='Period', data=combined_sales
)

# Customize the plot
plt.title('Average Automobile Sales Per Vehicle Type During Recession vs Non-Recession Periods', fontsize=14)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Average Automobile Sales', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Period')

# Display the plot
plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# Filtering data for non-recession periods
non_recession_data = df[df['Recession'] == 0]

# Aggregate sales and seasonality weight for each month
monthly_sales_data = non_recession_data.groupby('Month').agg(
    Automobile_Sales=('Automobile_Sales', 'sum'),
    Seasonality_Weight=('Seasonality_Weight', 'mean')
).reset_index()

# Plotting the bubble plot
plt.figure(figsize=(12, 8))
bubble_plot = sns.scatterplot(
    data=monthly_sales_data,
    x='Month',
    y='Automobile_Sales',
    size='Seasonality_Weight',
    sizes=(100, 2000),
    color='blue',
    alpha=0.7,
)

# Adding title and labels
plt.title('Seasonality Impact on Automobile Sales', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Automobile Sales', fontsize=14)

# Customizing the x-axis with month names
plt.xticks(ticks=range(1, 13), labels=[
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
], rotation=45)

# Adding gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
````

 ### TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.
 <br>Analyze the effect of the unemployment rate on vehicle type and sales during the Recession Period
 #### You can create a lineplot and title the plot as 'Effect of Unemployment Rate on Vehicle Type and Sales'


<details><summary>Click here for a hint</summary>

<p>
    Filter out the data for recession period<br>
     Make use of lineplot() from seaborn and pass the relavent data</p>
</details>


````python
# Filter data for recession periods (Recession == 1)
recession_data = df[df['Recession'] == 1]

# Group by 'Date' and 'Vehicle_Type' and calculate total sales
sales_by_date_vehicle = recession_data.groupby(['Date', 'Vehicle_Type'])['Automobile_Sales'].sum().unstack()

# Reset index to have 'Date' as a column
sales_by_date_vehicle.reset_index(inplace=True)

# Plotting the line plot
plt.figure(figsize=(12, 6))

# Plot total sales for each vehicle type
for vehicle_type in sales_by_date_vehicle.columns[1:]:
    plt.plot(sales_by_date_vehicle['Date'], sales_by_date_vehicle[vehicle_type], label=vehicle_type)

# Plot the unemployment rate
plt.plot(recession_data['Date'], recession_data['Unemployment_Rate'], label='Unemployment Rate', color='black', linestyle='--', linewidth=2)

# Add labels and title
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales')
plt.xlabel('Date')
plt.ylabel('Sales / Unemployment Rate')
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
````
